```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
new_df=pd.read_csv(r'housingdata.csv')
```

```python
new_df.columns
```

```python
from sklearn.preprocessing import MinMaxScaler
Scaler=MinMaxScaler()
num_vars = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking','price']

new_df[num_vars] = Scaler.fit_transform(new_df[num_vars])
```

```python
x=new_df.drop('price',axis=1)
y=new_df['price']
```

```python

x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42)
```

```python
x_train.shape
```

```python
Input=x_train.iloc[:, 0].to_numpy().flatten()
Input.size
```

```python
Input=x_train.to_numpy()


def mse(Yexpected,Ypred):
    return np.mean((Yexpected - Ypred) ** 2) 

Weights=np.random.randn(Input.shape[1])*0.1
Bias=np.random.randn()*0.1

#updatingg values for weights and biases


Yexpected=y_train.flatten()
# print(Input,Yexpected)
n=0



epochs=150
alpha=0.2
for i in range(epochs):
    Ypred=Input @ Weights + Bias
    dw = (-2 / len(Input)) * (Input.T @ (Yexpected - Ypred)) 
    db = (-2 / len(Input)) * np.sum(Yexpected - Ypred)
    Weights=Weights-alpha*dw
    Bias=Bias-alpha*db

    print("Updated Value")
    print (Yexpected[6],Ypred[6])
    MSE=mse(Yexpected,Ypred)
    print('MSE = ', MSE)
    print (Yexpected[0],Ypred[0])
    print('MSE = ', MSE)
    r21 = np.sum((Yexpected - Ypred)**2)
    r22 = np.sum((Yexpected - np.mean(Yexpected))**2)

    R2 = 1 - (r21 / r22)
    print("R² Accuracy:", R2 * 100, "%")
```

```python
Input=x_test.iloc[0]
Youtput=Weights * Input + Bias
print('Ouput Predicted :' , Youtput.iloc[0] ,'Output Expected :' ,y_test.iloc[0] )
```
